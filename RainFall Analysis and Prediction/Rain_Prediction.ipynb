## Take MAY,JUN,JUL as training set and predict for the AUG.

```python
df.head(0)
```

```python
predictors = []
for element in df.head(0):
    if element in ['STATE_UT_NAME','DISTRICT','MAY','JUN','JUL']:
        predictors.append(element)
print(predictors)
```

```python
target = ['AUG']
```

```python
# Separating out the predictors
x  = df.loc[:, predictors].values
# Separating out the target
y = df.loc[:,target].values
```

## Split the data into train data(2/3) and test data(1/3)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

## Deploy SVM method for prediction

```python
from sklearn import svm
from sklearn.datasets import make_classification
clf = svm.SVC()
X_train, y_train = make_classification()
X_test, y_test = make_classification()
clf.fit(X_train, y_train)
clf.predict(X_test)
clf.score(X_test, y_test)
```
